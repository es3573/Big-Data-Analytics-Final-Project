```python
#GLM_VIF Error 

import numpy as np
import matplotlib.pyplot as plt

N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

yvals = [0.0759306, 0.0572915, 0.060942,0.0527636,0.0754152,0.0759306] #1.	Unstacked Unbalanced 
rects1 = ax.bar(ind, yvals, width, color='r')
zvals = [0.24391,0.131852,0.0907223,0.0585648,0.226217,0.191195] #2.	Unstacked Oversampled 
rects2 = ax.bar(ind+width, zvals, width, color='g')
kvals = [0.258644,0.178328,0.180398,0.175812,0.242695,0.201258] # 2.	Unstacked Undersampled 
rects3 = ax.bar(ind+width*2, kvals, width, color='b')

ax.set_title("GLM_VIF")
ax.set_ylabel('Test Error')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('Logistic Regression', 'Decision Tree', 'Random Forest','Gradient Boost tree','Linear SVC','Multi layer Perceptron') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('Imbalanced data', 'Oversampled Data', 'UndrSampled Data') )

def autolabel(rects,labels):
    for rect, label in zip(rects, labels):
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.01*h, label,
                ha='center', va='bottom')

autolabel(rects1,yvals)
autolabel(rects2,zvals)
autolabel(rects3,kvals)

plt.show()

```
